```python
def temperature (c_temp,f_temp):
    cel_temp = (f_temp - 32) * 5 / 9 
    far_temp = (c_temp * 9 / 5) + 32 
    return cel_temp ,  far_temp

print(temperature(15,95))
```

```python
items = {}

def total_items():
    while True:
        item_name = input("Enter item name (or type 'exit' to stop): ")
        if item_name.lower() == 'exit':
            break
            
        try:
            item_price = int(input("Enter item price: "))
            items[item_name] = item_price
            
            print("Current items:", list(items.keys()))
            print("Total amount is:", sum(items.values()))
            print("-" * 20)
        except ValueError:
            print("Oops! Please enter a whole number for the price.")

total_items()
```
