```python
device = input("Enter state of Device (active or not_active): ")
if device == "not_active":
    print("device off")
elif device == "active":
    temperature = int(input("Enter Temperature: "))
    if temperature > 35:
        print("high temperature alert")
    elif temperature < 35:
        print("Cold temperature")
    else:
        print("Normal Temperature")
else:
    print("device off")
```

```python
order_amount = int(input("Enter Order Amount: "))
print(f"Order Amount:  {type (order_amount)}")
delivery_fee = 0 if order_amount > 300 else 30
print(f"delivery fee is: {delivery_fee}")
```

```python
seat = input("Enter seat type (Luxary/Sleeper/AC/General): ").lower()

match seat:
    case "ac":
        print("Ac seat- good")
    case "sleeper":
        print("Excellent")
    case "general":
        print("no Ac normal")
    case "luxary":
        print("Expensive and Super Excellent")
    case "-":
        print("Unknown")
```
